## Valutazione del tempo medio di risposta al variare della dimensione dei Batch

```python
dati = {
            'K' : ['64','64','64','64','64','64','64','64'],
           'B': ['0','64', '128', '256', '512','1024','2048','4096'],
           'wait': [45.561089,29.634315,29.561872,30.898281 ,30.175526,  31.219218,32.082667,31.955978  ],
           'somma': [ 45.561089 + 9.604199,29.634315 + 3.595397,29.561872 + 3.036612,30.898281 + 2.195099,30.175526 + 1.699956, 31.219218 + 1.375955,32.082667 + 1.781379,31.955978 + 1.005845],
           'sottrazione': [ 45.561089 - 9.604199,29.634315 - 3.595397,29.561872 - 3.036612,30.898281 - 2.195099,30.175526 - 1.699956, 31.219218 - 1.375955,32.082667 - 1.781379,31.955978 - 1.005845],
           'autocorrelazione' : [0.3,0.6,0.3,0.1,0.0,0.0,0.1,0.1]}


df=pd.DataFrame(dati, columns=['K','B','wait', 'somma','sottrazione', 'autocorrelazione'])
```

```python
import matplotlib.pyplot as plt
x=df.B

plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
colors=['orange', 'purple', 'green','red']
plt.figure(figsize=(20, 9))
plt.gca().set_prop_cycle(color=colors)
plt.plot(x,df.wait,"o--",label="Tempo medio di risposta",linewidth=2.5)
plt.plot(x,df.somma,label="Limite superiore",linewidth=2.5)
plt.plot(x,df.sottrazione,label="Limite inferiore", color='b',linewidth=2.5)

plt.axhline(y=31.880974, color='g', linestyle='-')
plt.title("Tempo di risposta centro vaccinale con k = 64",fontsize=30)
plt.xlabel("B",fontsize=20)
plt.ylabel("Wait",fontsize=20)
plt.legend(fontsize = 12)
plt.show()
```
