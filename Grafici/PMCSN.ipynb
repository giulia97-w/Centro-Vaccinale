## Valutazione del tempo medio di risposta al variare della dimensione dei Batch

```python
dati = {
            'K' : ['64','64','64','64','64','64','64'],
           'B': ['64', '128', '256', '512','1024','2048','4096'],
           'wait': [25.324572 ,27.842129   ,28.832612  ,30.098760  ,30.474565 ,30.703029   ,31.497895 ],
           'somma': [25.324572 + 2.573376 , 27.842129 + 2.743785,28.832612 + 2.312204 ,30.098760 + 1.876704 ,30.474565 + 1.404309,30.703029 + 1.378139 ,31.497895 + 1.161821 ],
           'sottrazione': [ 25.324572 - 2.573376, 27.842129 - 2.743785, 28.832612 - 2.312204 ,30.098760 - 1.876704,30.474565 - 1.404309,30.703029 - 1.378139,31.497895 - 1.161821]}


df=pd.DataFrame(dati, columns=['K','B','wait', 'somma','sottrazione'])
```

```python
df
```

```python
x=df.B

plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
colors=['orange', 'purple', 'green','red']
plt.figure(figsize=(20, 9))
plt.gca().set_prop_cycle(color=colors)
plt.plot(x,df.wait,"o--",label="Media",linewidth=2.5)
plt.plot(x,df.somma,label="Limite superiore",linewidth=2.5)
plt.plot(x,df.sottrazione,label="Limite inferiore", color='b',linewidth=2.5)

plt.axhline(y=31.320188, color='g', linestyle='-')
plt.title("Tempo di servizio centro vaccinale con k = 64",fontsize=30)
plt.xlabel("B",fontsize=20)
plt.ylabel("Wait",fontsize=20)
plt.legend(fontsize = 12)
plt.show()
```
